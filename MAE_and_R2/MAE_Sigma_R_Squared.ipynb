# 前言  

在資料科學領域中，機器學習模型的評估指標，是評量模型表現的重要指標。在執行一個資料科學專案時，我們對於建模前評估指標的設定，如何制定出一個合理且適切的標準，是一個關鍵且重要的決策過程。  
下列我們將針對機器學習模型的解釋能力指標 $R^2$ 與預測能力指標 $MAE$、$RMSE$，探討兩能力指標的相對應大小關係，以協助大家在訂定指標的標準時，能有明確的概念而訂定出更具體適用的指標標準。

# 評估指標的定義

假設 $y_i$ 為實際觀測值且 $\hat{y_i}$ 為一個 ML 模型的預測值，下列為常見的 ML 模型預測能力指標與解釋能力指標 :  

* 平均絕對誤差 Mean Absolute Error, MAE
$$\displaystyle \text{MAE} = \frac{1}{N} \sum_{i=1}^{N} |y_i-\hat{y_i}|$$  

\
* 均方誤差 Mean Square Error, MSE
$$\displaystyle \text{MSE} = \frac{1}{N} \sum_{i=1}^{N} (y_i-\hat{y_i})^2$$

\
* 均方根誤差 Root Mean Square Error, RMSE
$$\displaystyle \text{RMSE} = \sqrt{\frac{1}{N} \sum_{i=1}^{N} (y_i-\hat{y_i})^2}$$

\
* 決定係數 Coefficient of Determination, $R^2$
$$\displaystyle R^2 = 1 - \frac{\displaystyle \sum_{i=1}^{N} (y_i-\hat{y_i})^2}{\displaystyle \sum_{i=1}^{N} (y_i-\bar{y})^2 }$$

# $MAE$、$RMSE$、$R^2$ 與 $\sigma$ 的關係    



## <font color="#00dd00">**$\displaystyle MAE \leq RMSE$**</font>

【推導】: </br>    
藉由 [Cauchy-Schwarz Inequality](https://en.wikipedia.org/wiki/Cauchy%E2%80%93Schwarz_inequality)，對於所有的 $a_i \geq 0, 1\leq i\leq N$，
我們可知 \\
$$\displaystyle \frac{1}{N} \times \bigg(\sum_{i=1}^{N} a_i \bigg)^2 \leq \sum_{i=1}^{N} a_{i}^2 \text{ ，}$$

\
因此，  
\begin{aligned}\displaystyle
MAE^2 &= \frac{1}{N^2} \times \bigg(\sum_{i=1}^{N} |y_i-\hat{y_i}|\bigg)^2 \\
&& \\
&\leq \frac{1}{N} \times \sum_{i=1}^{N} (y_i-\hat{y_i})^2 \\
&& \\
&= RMSE^2 \text{ ，}
\end{aligned}  

也就是說，$MAE \leq RMSE$。

## <font color="#00dd00">**$\displaystyle \frac{MAE}{\sigma}\leq\sqrt{1-R^2}$**</font>

【推導】: </br>  
首先，從定義可知

\begin{aligned}\displaystyle
\bigg(\frac{\text{MAE}}{\sigma}\bigg)^2 &= \frac{\displaystyle \frac{1}{N^2} \times \bigg(\sum_{i=1}^{N} |y_i-\hat{y_i}| \bigg)^2} {\displaystyle \frac{1}{N} \times \sum_{i=1}^{N} (y_i-\bar{y})^2} \\
&& \\
&= \frac{1}{N} \times \frac{\bigg(\displaystyle \sum_{i=1}^{N} |y_i-\hat{y_i}| \bigg)^2}{\displaystyle \sum_{i=1}^{N} (y_i-\bar{y})^2} \text{ 。} \\
&& \\
\end{aligned}

另一方面，藉由 [Cauchy-Schwarz Inequality](https://en.wikipedia.org/wiki/Cauchy%E2%80%93Schwarz_inequality)，對於所有的 $a_i \geq 0, 1\leq i\leq N$，
我們可得 \\

$$\displaystyle \frac{1}{N} \times \bigg(\sum_{i=1}^{N} a_i \bigg)^2 \leq \sum_{i=1}^{N} a_{i}^2$$

\
所以，另外可得知

$$\frac{1}{N} \times \bigg(\displaystyle \sum_{i=1}^{N} |y_i-\hat{y_i}| \bigg)^2 \leq \sum_{i=1}^{N} |y_i-\hat{y_i}|^2 \text{ 。}$$
\
因此，最終我們可推得  
$$\bigg(\frac{\text{MAE}}{\sigma}\bigg)^2 \leq \frac{\displaystyle \sum_{i=1}^{N} |y_i-\hat{y_i}|^2}{\displaystyle \sum_{i=1}^{N} (y_i-\bar{y})^2} = 1-R^2 \text{ ,}$$
\
也就是說，
$$\displaystyle \frac{\text{MAE}}{\sigma} \leq \sqrt{1-R^2} \text{ 。}$$

## <font color="#00dd00">**$\displaystyle \frac{RMSE}{\sigma}=\sqrt{1-R^2}$**</font>

【推導】: </br>  
從定義可知

\begin{aligned}\displaystyle
\bigg(\frac{\text{RMSE}}{\sigma}\bigg)^2 &= \frac{\displaystyle \frac{1}{N} \times \sum_{i=1}^{N} (y_i-\hat{y_i})^2} {\displaystyle \frac{1}{N} \times \sum_{i=1}^{N} (y_i-\bar{y})^2} \\
&& \\
&= 1-R^2 \text{ ，} \\
&& \\
\end{aligned}
\
也就是說，
$$\displaystyle \frac{\text{RMSE}}{\sigma} = \sqrt{1-R^2} \text{ 。}$$

# $MAE\text{/}\sigma$ 的最大上限

- 若一 ML 模型的解釋能力指標 $R^2$ 要求須滿足 $R^2 \geq r_0$，則，此一 ML 模型的 $\displaystyle \frac{MAE}{\sigma}$ 的最大範圍為 $\big[0, \sqrt{1-r_0}\big]$。</br>

</center>
  
- 當一 ML 模型在測試集的 $\displaystyle \frac{MAE}{\sigma_{測試集}} \leq 0.3162$ 時，</br>
則，此一 ML 模型在測試集的解釋能力必滿足 $R^2 \geq 0.9$ 的要求。  
</br>
  
- 若我們想訓練出一 ML 模型，它在測試集的解釋能力必須滿足 $R^2 \geq 0.9$，</br>
則，此一 ML 模型在測試集的 $\displaystyle \frac{MAE}{\sigma_{測試集}}$ 不可大於 0.3162，也就是說，預測能力指標 $MAE$ 不可大於 $\sigma_{測試集} \times 0.3162$。  
</br>
  
- 當一 ML 模型的 $\displaystyle \frac{MAE}{\sigma} \geq 1$，則，此一 ML 模型的解釋能力指標必定 $R^2 \leq 0 $。

【證明】: 若解釋能力指標 $R^2$ 要求須滿足 $R^2 \geq r_0$，則，$\displaystyle \frac{MAE}{\sigma}$ 的最大範圍為 $\big[0, \sqrt{1-r_0}\big]$。</br>  
假設 $\displaystyle \frac{MAE}{\sigma} > \sqrt{1-r_0}$。則，我們可得知
$$1-R^2 \geq \bigg(\frac{MAE}{\sigma}\bigg)^2 > 1-r_0 \text{ , }$$ </br>  
因此，我們可得 $ R^2 < r_0$ ，與要求條件 $R^2 \geq r_0$ 矛盾。

# Python 範例程式

## 載入套件

```python
import sys
import os

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.font_manager import fontManager
import sklearn
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
# 繪圖繁中字體設定
!wget -O TaipeiSansTCBeta-Regular.ttf https://drive.google.com/uc?id=1eGAsTN1HBpJAkeVM57_C7ccp7hbgSz3_&export=download

fontManager.addfont('TaipeiSansTCBeta-Regular.ttf')
mpl.rc('font', family='Taipei Sans TC Beta')

print('Python version:', sys.version[0:7])
print('NumPy version:', np.__version__)
print('Pandas version:', pd.__version__)
print('Matplotlib version:', mpl.__version__)
print('Scikit-Learn version:', sklearn.__version__)
```

## 定義函數

```python
def square_root_of_1_minus_R2(r2):
    '''
    計算 (1-R2)^(0.5) 。
    '''

    return np.sqrt(1-r2)
```

```python
def calculate_range(r2_min, r2_max, size=1000):
    '''
    給定 R2 的上下界，計算 MAE/Sigma 的上下界。
    '''

    range_of_r2 = np.linspace(r2_min, r2_max, size)
    range_mae_over_sigma = square_root_of_1_minus_R2(range_of_r2)

    return range_of_r2, range_mae_over_sigma
```

## $\frac{MAE}{\sigma}$ 與 $R^2$ 的關係示意圖

```python
plt.figure(figsize=(9, 6))

# 座標軸
plt.arrow(-1.5, 0, 3, 0, color='black', linewidth=2,
          head_width=0.05, head_length=0.03, length_includes_head=True)
plt.arrow(0, 0, 0, 1.75, color='black', linewidth=2,
          head_width=0.05, head_length=0.03, length_includes_head=True)

# Curve : MAE/Sigma = (1-R^2)^(0.5)
x = np.linspace(-1.5, 1, 10000)
y_limit = np.sqrt(1-x)
plt.plot(x, y_limit, ls='--', color='black')

# R^2 >= r0 = 0.7
x_interval, y_interval = calculate_range(r2_min=0.7, r2_max=1)
plt.hlines(y=0, xmin=0.7, xmax=1, color='green', linewidth=3)
plt.vlines(x=0, ymin=square_root_of_1_minus_R2(0.7), ymax=square_root_of_1_minus_R2(1), color='green', linewidth=3)
plt.arrow(0.7, square_root_of_1_minus_R2(0.7), -0.7, 0, ls=':', color='green', linewidth=2,
          head_width=0.05, head_length=0.03, length_includes_head=True)
plt.vlines(x=0.7, ymin=0, ymax=square_root_of_1_minus_R2(0.7), linestyles=':', color='green', linewidth=2)
plt.fill_between(x_interval, y_interval, color='green', alpha=0.2,
                 label=r'$R^2\geq r_0 $'+' 且\n'+'$MAE/\sigma\leq \sqrt{1-r_0}$')

# 說明文字
plt.text(-1.2, 1.5, '$\\frac{MAE}{\sigma}=(1-R^2)^{1/2}$', fontsize=12)
plt.text(0.65, -0.075, '$r_0$', color='green', fontsize=12)
plt.text(-0.35, 0.525, '$\sqrt{1-r_0}$', color='green', fontsize=12)

plt.legend()
plt.xlabel('$R^2$', fontsize=12)
plt.ylabel('$MAE/\sigma$', fontsize=12)
plt.title('在 '+'$R^2 \geq r_0$ 的條件中'+' \n'+'$MAE/\sigma$ 的相對範圍', fontsize=14)
plt.grid(False)

img_path = 'MAE_Sigma_R2.png'
plt.savefig(img_path, bbox_inches='tight', format='png')

plt.show()
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(121)
# 座標軸
plt.arrow(0, 0, 1.2, 0, color='black', linewidth=2,
          head_width=0.05, head_length=0.03, length_includes_head=True)
plt.arrow(0, 0, 0, 1.2, color='black', linewidth=2,
          head_width=0.05, head_length=0.03, length_includes_head=True)

# Curve : MAE/Sigma = (1-R^2)^(0.5)
x = np.linspace(0, 1, 10000)
y_limit = np.sqrt(1-x)
plt.plot(x, y_limit, ls='--', color='black')
plt.text(0.1, 1, '$\\frac{MAE}{\sigma}=(1-R^2)^{1/2}$', fontsize=12)

# 訓練集中，模型解釋能力的標準: R^2 >= r0 = 0.7
x_interval, y_interval = calculate_range(r2_min=0.7, r2_max=1)
plt.hlines(y=0, xmin=0.7, xmax=1, color='green', linewidth=3)
plt.vlines(x=0, ymin=square_root_of_1_minus_R2(0.7), ymax=square_root_of_1_minus_R2(1), color='green', linewidth=3)
plt.vlines(x=0.7, ymin=0, ymax=square_root_of_1_minus_R2(0.7), ls='--', color='green', alpha=0.7)
plt.arrow(0.7, square_root_of_1_minus_R2(0.7), -0.7, 0, ls=':', color='green', alpha=0.7, linewidth=2,
          head_width=0.05, head_length=0.03, length_includes_head=True)
plt.fill_between(x_interval, y_interval, color='green', alpha=0.3,
                 label=r'$R^2 \geq r_0$'+' 且\n '+'$MAE/\sigma\leq \sqrt{1-r_0}$')
plt.text(0.7, -0.05, '$r_0$', color='green', fontsize=12)
plt.text(0.025, 0.6, '$\sqrt{1-r_0}$', color='green', fontsize=12)
plt.legend()
plt.xlabel('$R^2$', fontsize=12)
plt.ylabel('$MAE/\sigma$', fontsize=12)
plt.title('訓練集中 ML 模型指標可達成範圍', fontsize=14)

plt.subplot(122)
# 座標軸
plt.arrow(0, 0, 1.2, 0, color='black', linewidth=2,
          head_width=0.05, head_length=0.03, length_includes_head=True)
plt.arrow(0, 0, 0, 1.2, color='black', linewidth=2,
          head_width=0.05, head_length=0.03, length_includes_head=True)

# Curve : MAE/Sigma = (1-R^2)^(0.5)
x = np.linspace(0, 1, 10000)
y_limit = np.sqrt(1-x)
plt.plot(x, y_limit, ls='--', color='black')
plt.text(0.1, 1, '$\\frac{MAE}{\sigma}=(1-R^2)^{1/2}$', fontsize=12)

# 訓練集中，ML 模型指標的評估準則: R^2 >= r0 = 0.7
x_interval, y_interval = calculate_range(r2_min=0.7, r2_max=1)
plt.hlines(y=0, xmin=0.7, xmax=1, color='green', linewidth=3)
plt.vlines(x=0, ymin=square_root_of_1_minus_R2(0.7), ymax=square_root_of_1_minus_R2(1), color='green', linewidth=3)
plt.hlines(y=square_root_of_1_minus_R2(0.7), xmin=0, xmax=0.7, ls='--', color='green', alpha=0.2)
plt.vlines(x=0.7, ymin=0, ymax=square_root_of_1_minus_R2(0.7), ls='--', color='green', alpha=0.2)
plt.fill_between(x_interval, y_interval, color='green', alpha=0.3)

# 測試集中，ML 模型指標的評估準則: R^2 >= 0.9*r0 = 0.9*0.7
x_interval, y_interval = calculate_range(r2_min=0.9*0.7, r2_max=1)
plt.hlines(y=0, xmin=0.9*0.7, xmax=1, color='red', linewidth=3)
plt.vlines(x=0, ymin=square_root_of_1_minus_R2(0.9*0.7), ymax=square_root_of_1_minus_R2(1), color='red', linewidth=3)
plt.arrow(0.9*0.7, square_root_of_1_minus_R2(0.9*0.7), -0.9*0.7, 0, ls=':', color='red', alpha=0.7, linewidth=2,
          head_width=0.05, head_length=0.03, length_includes_head=True)
plt.vlines(x=0.9*0.7, ymin=0, ymax=square_root_of_1_minus_R2(0.9*0.7), ls=':', color='red', alpha=0.7, linewidth=2)
plt.fill_between(x_interval, y_interval, color='red', alpha=0.3,
                 label=r'$R^2 \geq c\times r_0$'+' 且\n'+'$MAE/\sigma\leq \sqrt{1-c \\times r_0}$')
plt.text(0.55, -0.065, '$c\\times r_0 $'+', for some $c\in(0, 1]$', color='red', fontsize=12)
plt.text(0.025, 0.65, '$\sqrt{1-c\\times r_0}$', color='red', fontsize=12)
plt.legend()
plt.xlabel('$R^2$', fontsize=12)
plt.title('測試集中 ML 模型指標的訂定範圍', fontsize=14)

img_path = 'Criteria_of_Training_and_Testing_Dataset.png'
plt.savefig(img_path, bbox_inches='tight', format='png')

plt.show()
```

## 範例資料

```python
sample_data_1 = pd.read_excel(r'.//sample_data_for_MAE_Sigma_R2.xlsx', sheet_name='工作表1')
y_truth = sample_data_1['Ground_Truth']
y_pred = sample_data_1['Prediction']

truth_std = np.std(y_truth)
r2 = r2_score(y_truth, y_pred)
mae = mean_absolute_error(y_truth, y_pred)
rmse = mean_squared_error(y_truth, y_pred, squared=False)
mae_over_std = np.round(mae/truth_std, 4)
rmse_over_std = np.round(rmse/truth_std, 4)
```

```python
plt.figure(figsize=(12, 5))
plt.plot(y_truth, color='black', label='Ground Truth\n$\sigma =$'+str(np.round(truth_std, 4)))
plt.plot(y_pred, color='red', label='Prediction\nMAE ='+str(np.round(mae, 4))+'\nRMSE ='+str(np.round(rmse, 4)))
plt.ylim(bottom=1150, top=1280)
plt.legend()
plt.xlabel('Index', fontsize=12)
plt.title('Prediction and Ground Truth', fontsize=12)

# R2 與 MAE/Sigma 的資訊說明
text = '$R^2=$'+str(np.round(r2, 4))+'\n$MAE/\sigma=$'+str(mae_over_std)+'\n$RMSE/\sigma=$'+str(rmse_over_std)
plt.text(150, 1255, text, fontsize=12)

img_path = 'Demo_1_of_MAE_Sigma_R2.png'
plt.savefig(img_path, bbox_inches='tight', format='png')

plt.show()
```
