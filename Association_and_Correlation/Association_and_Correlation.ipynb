 # 前言  

在資料分析的過程中，探索變數之間的關聯性/相關性，除能有助於我們理解數據中潛在或顯著的關係外，也能輔助我們在訓練機器學習模型時，尋找到具有影響力且重要的特徵變數，以提升模型的預測能力與解釋能力。  
下列我們將針對不同的變數型態，介紹各自適用的統計檢定方法，包括 : Contingency Coefficient、Cramer’s V、Pearson 相關係數、Spearman 與 Kendall 等級相關係數等，以讓大家能更精確地選用合適的統計工具，來衡量變數之間的關係。除此之外，最後還將介紹一個新的係數指標 : Chatterjee 相關係數 $\xi$，它可用來評估兩變數之間是否存在某個函數的對應關係。

# 檢定方法

| 變數 $X_1$ 的資料型態</br></br> | 變數 $X_2$ 的資料型態</br></br> | 關聯性/相關性檢定</br></br> |    
|:---:|:---:|:---|    
| 名目/次序變數 | 名目/次序變數 | Pearson's Contingency Coefficient、</br></br>Cramer’s V、Tschuprow’s T |

</center>
</br>  

其中，類別型變數(Categorical Variable)可再細分成下列兩類 :  
* 名目變數(Nominal Variable) : 不同的數值代表不同類型的事物(/型態)，例如: 性別、品牌、顏色等；     
* 次序變數(Ordinal Variable) : 數值的大小能詮釋某種特性的排序，例如: 衣服尺寸、滿意度等。

# <font color="#00dd00">**Pearson's Contingency Coefficient**</font>、<font color="#00dd00">**Cramer’s V**</font>  與 <font color="#00dd00">**Tschuprow’s T**</font>    

* Pearson Contingency Coefficient、Cramer’s V 與 Tschuprow’s T 三者皆為用來衡量兩名目(或次序)類別型變數之間的關聯性(Association)。  

* 檢定統計量  
  * Pearson's Contingency Coefficient :  
     \begin{aligned}
     C &\equiv \sqrt{\displaystyle \frac{\chi^2}{\chi^2+N}} \in [0, 1)\\ \\
     \end{aligned}  
  * Cramer’s V :  
     \begin{aligned}
     V &\equiv \sqrt{\displaystyle \frac{\chi^2 /N}{\min \big\{k-1, r-1 \big\} }} \in [0, 1]\\ \\
     \end{aligned}  
  * Tschuprow’s T :  
     \begin{aligned}  
     T &\equiv \sqrt{\displaystyle \frac{\chi^2 /N}{(k-1)(r-1) }} \in [0, 1]\\ \\
     \end{aligned}

  其中，$\chi^2$ 為[卡方統計量](https://en.wikipedia.org/wiki/Chi-squared_test)，$N$ 為兩變數成對的樣本數量，$k$ 與 $r$ 分別為兩變數的類別數量。  
  **注意 :** 當存有任何一個類別的樣本數過少($<5$)時，卡方統計量 $\chi_2$ 的計算結果會偏高，易造成關聯性高估的現象。因此，為確保各檢定統計量的可信度，建議每個類別中至少有 5 個樣本數據。另外，當存有某些類別樣本數不足 5 時，建議可將高相似性的類別作合併，以符合理論的假設條件。

* Pearson's Contingency Coefficient、Cramer’s V 與 Tschuprow’s T 皆提供兩類別型變數間的關聯強度，0 皆表示兩變數完全無關聯，越近似 1 皆表示兩變數關聯性越高，然而，<font color="#FFFF00">**關聯性高不可直接推論兩變數具有因果關係**</font>。  

* Pearson's Contingency Coefficient、Cramer’s V 與 Tschuprow’s T 關聯強弱的閾值(Threshold)無統一的標準，需視研究領域而定。對於一些著重於相關性較強的研究領域，例如 : 物理、化學、醫學、生物學等研究，較高的標準對於研究人員會較具意義，例如 : 在醫療臨床試驗的研究 [1] 中，$V\geq 0.5$ 為高度相關、$0.3 \leq V <0.5$ 為中度相關，然而，在社會科學的研究中 [2] ，$V > 0.15$ 才視為強相關。  </br></br>
參考文獻 :   
  * **[1]** Kakudji, B. K., Mwila, P. K., Burger, J. R., & Plessis, J. M. D. (2020). [Epidemiological, clinical and diagnostic profile of breast cancer patients treated at Potchefstroom regional hospital](https://www.ajol.info/index.php/pamj/article/view/210808), South Africa, 2012-2018: An open-cohort study. Pan African Medical Journal, 36(1), Article 1.  
  * **[2]** Chanvril-Ligneel, F., & Hay, V. L. (2014). Méthodes Statistiques pour les Sciences Sociales. ELLIPSES.  
</br>
* 資料情境的選用 :  
  * 當兩變數的類別數量相差很大時，Cramer's V 相較於 Tschuprow’s T 適用，因 Cramer's V 主要受到類別數量較少的變數所影響；  
  * 當兩變數的類別皆很多類時，Tschuprow’s T 相較於 Cramer's V 適用，因 Tschuprow’s T 會同時考量兩變數的類別數量；
  * Pearson's Contingency Coefficient 不受類別數量所影響；  
  * 當類別資料的分布均勻時，三者皆適用；
  * 當類別資料的分布不均勻時，Tschuprow’s T 相對較適用且穩健。

````python
#@title Python 範例程式碼

#@markdown 在運用 Python 計算 Pearson's Contingency Coefficient、Cramer’s V 與 Tschuprow’s T 統計量時，我們需先透過 [```pandas.crosstab(X1, X2)```](https://pandas.pydata.org/pandas-docs/version/2.0/reference/api/pandas.crosstab.html#pandas-crosstab) 函數，取得兩類別型變數 $X_1$ 與 $X_2$ 的列聯表(Contingency Table)，
#@markdown 接續，使用 [```scipy.stats.contingency.association()```](https://docs.scipy.org/doc/scipy-1.11.4/reference/generated/scipy.stats.contingency.association.html#scipy-stats-contingency-association) 函數從聯表中計算出 Pearson's Contingency Coefficient、Cramer's V 與 Tschuprow’s T 統計量。

#@markdown 下面範例將利用[汽車評估資料集](https://archive.ics.uci.edu/dataset/19/car+evaluation)中的類別型特徵變數 : 安全性(safety)與目標變數 : 汽車等級(class)，計算 Cramer's V 統計量約 0.372、Tschuprow’s T 統計量約 0.337 與 Pearson's C 統計量約 0.466，另外，從其堆疊長條圖可觀察到，最佳等級(vgood)車輛的安全性皆為最高，反之，最差等級(unacc)車輛約近半數為低安全性，低安全性的車輛皆屬於最差等級。

import sys
import os

import scipy
from scipy.stats.contingency import association
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

print('Python version:', sys.version[0:7])
print('SciPy version:', scipy.__version__)
print('NumPy version:', np.__version__)
print('Pandas version:', pd.__version__)
print('Matplotlib version:', mpl.__version__)
print('Seaborn version:', sns.__version__)
print('===================\n' )

# 讀取汽車評估資料集(https://archive.ics.uci.edu/dataset/19/car+evaluation)
car_data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/car/car.data', header=None)
car_data.columns = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']

# 選取特徵欄位
X = car_data['safety']
Y = car_data['class']

# 計算兩類別型特徵的列聯表(Contingency Table)
contingency_table = pd.crosstab(Y, X, margins=False)

# 計算 Pearson's Contingency Coefficient、Cramer’s V 與 Tschuprow’s T 統計量
c = association(contingency_table.to_numpy(), method='pearson')
v = association(contingency_table.to_numpy(), method='cramer')
t = association(contingency_table.to_numpy(), method='tschuprow')

# 資料視覺化 - 列聯表熱力圖
plt.figure(figsize=(7, 6))
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='OrRd')
plt.title("Contingency Table\n(Cramer's V = "+str(np.round(v, 3))+", Tschuprow's T = "+str(np.round(t, 3))+", Pearson's C = "+str(np.round(c, 3))+")",
          fontsize=10)

img_path = os.path.join(os.getcwd(), 'Pearson_Cramer_and_Tschuprow_with_Car_Evaluation_Dataset.png')
plt.savefig(img_path, bbox_inches='tight', format='png')

plt.show()

print(' ')

# 資料視覺化 - 堆疊長條圖
barplot = contingency_table.plot(kind='bar', stacked=True, rot=0, figsize=(6.5, 6))
barplot.set_title('Bar Plot \nwith safety and class')
img_path = os.path.join(os.getcwd(), 'Barplot_with_Car_Evaluation_Dataset.png')
plt.savefig(img_path, bbox_inches='tight', format='png')

plt.show()
````

# <font color="#00dd00">**Pearson Correlation Coefficient**</font>  

* Pearson 相關係數 $\rho$ 用來衡量兩連續型特徵之間，是否具有線性關係(Linear Relation)。

* 數學定義  
\begin{aligned}
\rho_{X, Y} &\equiv \frac{\displaystyle \sum_{i=1}^{N}(x_i-\bar x)(y_i-\bar y)}{\sqrt{\displaystyle \sum_{i=1}^{N}(x_i-\bar x)^2 \sum_{i=1}^{N}(y_i-\bar y)^2}}\\ \\
&= \frac{cov(X, Y)}{\sigma_X \sigma_Y} \in [-1, 1]\\ \\
\end{aligned}
其中，$\bar x$、$\bar y$ 分別為兩特徵變數 $X$ 與 $Y$ 的平均數(Mean)，$\sigma_X$、$\sigma_Y$ 分別為兩特徵變數 $X$ 與 $Y$ 的[標準差(Standard Deviation)](https://en.wikipedia.org/wiki/Standard_deviation)，$cov(X, Y)$ 為[共變異數(Covariance)](https://en.wikipedia.org/wiki/Covariance)。

* Pearson 相關係數的解讀   
  
<center>

| Pearson 相關係數 | 兩特徵變數 $X$ 與 $Y$ 的線性關係|   
|:---|---:|    
| $0.9 \leq |\rho| \leq 1$ | 極高度線性關係 |  
| $0.7 \leq |\rho| < 0.9$ | 高度線性關係 |
| $0.5 \leq |\rho| < 0.7$ | 中度線性關係 |  
| $0.3 \leq |\rho| < 0.5$ | 低度線性關係 |  

</center>

````python
#@title Python 範例程式碼

#@markdown 在 SciPy 套件中，我們可使用 [```scipy.stats.pearsonr(x, y)```](https://docs.scipy.org/doc/scipy-1.11.4/reference/generated/scipy.stats.pearsonr.html#scipy-stats-pearsonr) 函數來計算 Pearson 相關係數。
#@markdown 下面範例為在三種不同的特徵關係下， Pearson 相關係數的大小。

import sys
import os

import scipy
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

print('Python version:', sys.version[0:7])
print('SciPy version:', scipy.__version__)
print('NumPy version:', np.__version__)
print('Pandas version:', pd.__version__)
print('Matplotlib version:', mpl.__version__)
print('===================\n' )

np.random.seed(42)

# 生成 3 組樣本數據
X = np.random.uniform(low=1, high=10, size=100)
Y1 = 3*X + 1.5*np.random.randn(100)
Y2 = X**2 + 30*np.random.randn(100)
Y3 = 0.8*np.sin(X) + 0.15*np.random.randn(100)

# 計算 Pearson 相關係數
res1 = scipy.stats.pearsonr(X, Y1)
res2 = scipy.stats.pearsonr(X, Y2)
res3 = scipy.stats.pearsonr(X, Y3)
pr1 = res1.statistic
pr2 = res2.statistic
pr3 = res3.statistic

# 資料視覺化
plt.figure(figsize=(12, 3))
plt.subplot(131)
plt.scatter(X, Y1, s=8)
plt.title('Pearsonr $\\rho$ =' +str(np.round(pr1, 3)))
plt.ylabel('Y', rotation=0)

plt.subplot(132)
plt.scatter(X, Y2, s=8)
plt.title('Pearsonr $\\rho$ =' +str(np.round(pr2, 3)))
plt.xlabel('X')

plt.subplot(133)
plt.scatter(X, Y3, s=8)
plt.title('Pearsonr $\\rho$ =' +str(np.round(pr3, 3)))

img_path = 'Pearsonr_Correlation_Coefficient.png'
plt.savefig(img_path, bbox_inches='tight', format='png')

plt.show()
````

# <font color="#00dd00">**Spearman's Rank Correlation Coefficient**</font>

* Spearman 等級相關係數 $r_s$ 係衡量兩連續型特徵之間，是否存有單調性(Monotonicity)，也就是說，當有兩個特徵 $X$ 與 $Y$ 時，較大的 $X$ 數值是否對應較大的 $Y$ 數值。  

* Spearman 等級相關係數的計算方法，與 Pearson 相關係數相同，但，在計算 Spearman 等級相關係數時，考慮的是特徵變數中的數值等序，也就是說，先將兩特徵變數中的數值，分別給予大到小的排序，再對這裡兩組排序計算 Pearson 相關係數。  

* 數學定義  
\begin{aligned}  
r_s &\equiv \rho_{R(X), R(Y)} \\ \\ &= \frac{cov\big(R(X), R(Y)\big)}{\sigma_{R(X)} \sigma_{R(Y)}} \in [-1, 1] \\ \\
\end{aligned}
其中，$\sigma_{R(X)}$、$\sigma_{R(Y)}$ 為兩特徵變數分別轉為有序變數(Ranked Variable)後的標準差(Standard Deviation)，$cov(R(X), R(Y))$ 為兩特徵變數分別轉換成有序變數後的共變異數。
  
* 當 $r_s$ 越趨近於 $1$ 時，表示兩特徵之間存有單調遞增的情況；反之，當 $r_s$ 越趨近於 $-1$ 時，表示兩特徵之間存有單調遞減的情況。






````python
#@title 離群值對於計算相關係數的影響

#@markdown 當樣本數據存有離群值(Outlier)時，因 Spearman 等級相關係數係藉由排序等級來計算相關係數，所以，相較於 Pearson 相關係數，Spearman 等級相關係數受離群值的影響較小。
#@markdown 因此，<font color="#FFFF00">當樣本數據存有少量的離群值時，建議使用 Spearman 等級相關係數，來評估兩特徵變數的相關性。</font>

import sys
import os

import scipy
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

print('Python version:', sys.version[0:7])
print('SciPy version:', scipy.__version__)
print('NumPy version:', np.__version__)
print('Pandas version:', pd.__version__)
print('Matplotlib version:', mpl.__version__)
print('===================\n' )

np.random.seed(42)

# 生成 2 組樣本數據
X1 = np.random.uniform(low=1, high=10, size=100)
Y1 = 3*X1 + np.random.randn(100)
X2 = np.append(np.random.uniform(low=1, high=10, size=95), np.random.uniform(low=15, high=18, size=5))
Y2 = np.append(3*X2[0:95]+np.random.randn(95), np.random.uniform(low=22, high=28, size=5))

# 計算 Pearsom 相關係數與 Spearmanr 等級相關係數
res1 = scipy.stats.pearsonr(X1, Y1)
res2 = scipy.stats.pearsonr(X2, Y2)
pr1 = res1.statistic
pr2 = res2.statistic

res1 = scipy.stats.spearmanr(X1, Y1)
res2 = scipy.stats.spearmanr(X2, Y2)
sr1 = res1.statistic
sr2 = res2.statistic

# 資料視覺化
plt.figure(figsize=(9, 3))
plt.subplot(121)
plt.scatter(X1, Y1, s=8)
plt.title('Pearson $\\rho$ =' +str(np.round(pr1, 3))+'\n'+'Spearman $r_s$ =' +str(np.round(sr1, 3)))
plt.xlabel('X')
plt.ylabel('Y', rotation=0)

plt.subplot(122)
plt.scatter(X2, Y2, s=8)
plt.title('Pearson $\\rho$ =' +str(np.round(pr2, 3))+'\n'+'Spearman $r_s$ =' +str(np.round(sr2, 3)))
plt.xlabel('X')

img_path = 'Pearson_and_Spearman_about_Outliers.png'
plt.savefig(img_path, bbox_inches='tight', format='png')

plt.show()
````

````python
#@title Python 範例程式碼

#@markdown 在 SciPy 套件中，我們可使用 [```scipy.stats.spearmanr(a, b)```](https://docs.scipy.org/doc/scipy-1.11.4/reference/generated/scipy.stats.spearmanr.html#scipy-stats-spearmanr)函數來計算 Spearman 等級相關係數。 下面範例為在三種不同的特徵關係下， Spearman 等級相關係數的大小。

import sys
import os

import scipy
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

print('Python version:', sys.version[0:7])
print('SciPy version:', scipy.__version__)
print('NumPy version:', np.__version__)
print('Pandas version:', pd.__version__)
print('Matplotlib version:', mpl.__version__)
print('===================\n' )

np.random.seed(42)

# 生成 3 組樣本數據
X = np.random.uniform(low=1, high=10, size=100)
Y1 = 3*X + 1.5*np.random.randn(100)
Y2 = X**2 + 30*np.random.randn(100)
Y3 = 0.8*np.sin(X) + 0.15*np.random.randn(100)

# 計算 Spearmanr 相關係數
res1 = scipy.stats.spearmanr(X, Y1)
res2 = scipy.stats.spearmanr(X, Y2)
res3 = scipy.stats.spearmanr(X, Y3)
r1 = res1.statistic
r2 = res2.statistic
r3 = res3.statistic

# 資料視覺化
plt.figure(figsize=(12, 3))
plt.subplot(131)
plt.scatter(X, Y1, s=8)
plt.title('Spearman $r_s$ =' +str(np.round(r1, 3)))
plt.ylabel('Y', rotation=0)

plt.subplot(132)
plt.scatter(X, Y2, s=8)
plt.title('Spearman $r_s$ =' +str(np.round(r2, 3)))
plt.xlabel('X')

plt.subplot(133)
plt.scatter(X, Y3, s=8)
plt.title('Spearman $r_s$ =' +str(np.round(r3, 3)))

img_path = 'Spearman_Rank_Correlation_Coefficient.png'
plt.savefig(img_path, bbox_inches='tight', format='png')

plt.show()
````

# <font color="#00dd00">**Kendall Rank Correlation Coefficient**</font>  

* Kendall 等級相關係數 $\tau$ 係衡量兩特徵變數之間，是否存有次序關聯(Ordinal Association)。  

* 數學定義 :   
  將兩特徵變數 $X$ 與 $Y$ 中的樣本數據，組成配對資料 $(x_i, y_i), \cdots, (x_j, y_j), \text{ for } 1\leq i<j \leq N$，且定義
  * [一致數對(Concordant Pairs)](https://en.wikipedia.org/wiki/Concordant_pair)表示，當 $i<j$ 時，$x_i<x_j$ 且 $y_i<y_j$ (或者， $x_i>x_j$ 且 $y_i>y_j$)；   
  * 不一致數對(Disconcordant Pairs)表示，當 $i<j$ 時，$x_i<x_j$ 且 $y_i>y_j$ (或者， $x_i>x_j$ 且 $y_i<y_j$)。  </br>
  
  * 若無 $x_i=x_j$ 或 $y_i=y_j$ 的情況，則， Kendall 等級相關係數的定義為   

    \begin{aligned} \displaystyle    
    \tau_a &\equiv \frac{\text{一致數對的總數 } - \text{不一致數對的總數}}{C_{2}^{N}} \in [-1, 1] \\ \\
    \end{aligned}  
  
  * 若存在 $x_i=x_j$ 或 $y_i=y_j$ 的情況，則， Kendall 等級相關係數的定義為

    \begin{aligned} \displaystyle    
    \tau_b &\equiv \frac{\text{一致數對的總數 } - \text{不一致數對的總數}}{\sqrt{\big(C_{2}^{N}-\sum_{i} \frac{t_i(t_i-1)}{2}\big)\big(C_{2}^{N}-\sum_{j} \frac{u_j(u_j-1)}{2}\big)}} \in [-1, 1] \\ \\
    \end{aligned}   

    其中，$t_i$ 與 $u_j$ 分別為特徵變數 $X$ 與 $Y$ 中排序相同的個數。    

  * 若存在 $x_i=x_j$ 或 $y_i=y_j$ ，且兩特徵變數的樣本數不同，則， Kendall 等級相關係數的定義為

    \begin{aligned} \displaystyle    
    \tau_c &\equiv 2\times\frac{\text{一致數對的總數 } - \text{不一致數對的總數}}{(n+m)^2\times \frac{\min\{n, m\}-1}{\min\{n, m\}}}  \in [-1, 1] \\ \\
    \end{aligned}   

    其中，$n\not= m$ 分別為兩特徵變數 $X$ 與 $Y$ 的樣本數。

* 當 $\tau$ 越趨近於 $1$ 時，表示兩特徵之間存有單調遞增的情況；反之，當 $\tau$ 越趨近於 $-1$ 時，表示兩特徵之間存有單調遞減的情況。

* Kendall 與 Spearman 等級相關係數的比較 :   
  * 一般而言，通常 Spearman $r_s$ 較 Kendall $\tau$ 大；
  * 一般而言，Kendall $\tau$ 較 Spearman $r_s$ 來得穩健，當樣本數小或存有離群值時，建議採用 Kendall $\tau$ 來評估單調性；
  * Kendall $\tau$ 的解釋較 Spearman $r_s$ 直觀，直觀可粗略解讀成，樣本數據的一致性與不一致性的機率差。

````python
#@title Python 範例程式碼

#@markdown 在 SciPy 套件中，我們可使用 [```scipy.stats.kendalltau(x, y, variant='b')```](https://docs.scipy.org/doc/scipy-1.11.4/reference/generated/scipy.stats.kendalltau.html#scipy-stats-kendalltau)函數來計算 Kendall 等級相關係數。 下面範例為在三種不同的特徵關係下， Kendall 等級相關係數的大小。

import sys
import os

import scipy
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

print('Python version:', sys.version[0:7])
print('SciPy version:', scipy.__version__)
print('NumPy version:', np.__version__)
print('Pandas version:', pd.__version__)
print('Matplotlib version:', mpl.__version__)
print('===================\n' )

np.random.seed(42)

# 生成 3 組樣本數據
X = np.random.uniform(low=1, high=10, size=100)
Y1 = 3*X + 1.5*np.random.randn(100)
Y2 = X**2 + 30*np.random.randn(100)
Y3 = 0.8*np.sin(X) + 0.15*np.random.randn(100)

# 計算 Kendall 等級相關係數
res1 = scipy.stats.kendalltau(X, Y1)
res2 = scipy.stats.kendalltau(X, Y2)
res3 = scipy.stats.kendalltau(X, Y3)
tau1 = res1.statistic
tau2 = res2.statistic
tau3 = res3.statistic

# 資料視覺化
plt.figure(figsize=(12, 3))
plt.subplot(131)
plt.scatter(X, Y1, s=8)
plt.title('Kendall $\\tau$ = ' +str(np.round(tau1, 3)))
plt.ylabel('Y', rotation=0)

plt.subplot(132)
plt.scatter(X, Y2, s=8)
plt.title('Kendall $\\tau$ = ' +str(np.round(tau2, 3)))
plt.xlabel('X')

plt.subplot(133)
plt.scatter(X, Y3, s=8)
plt.title('Kendall $\\tau$ = ' +str(np.round(tau3, 3)))

img_path = 'Kendall_Rank_Correlation_Coefficient.png'
plt.savefig(img_path, bbox_inches='tight', format='png')

plt.show()
````

# <font color="#00dd00">**Chatterjee's Coefficient Correlation $\xi$**</font>    

* Chatterjee 相關係數 $\xi$ 為 Sourav Chatterjee 在 2020 年發表的論文中所提出，它係衡量兩特徵變數之間，是否存有某種函數的對應關係，也就是說，$Y\approx f(X)$ 或 $X\approx f(Y)$。   

* 數學定義 :  
  首先，將兩特徵變數 $X$ 與 $Y$ 組成[有序對(Ordered Pair)](https://en.wikipedia.org/wiki/Ordered_pair)，且對 $X$ 中的數據進行由小到大的排序，也就是說，排序後我們可得 $(x_1, y_1), (x_2, y_2), \cdots, (x_N, y_N)$ 滿足 $x_1 \leq x_2 \leq \cdots \leq x_N$。
  * 在兩特徵變數 $X$ 與 $Y$ 中皆無重複的數據時，則，Chatterjee 相關係數 $\xi$ 的定義為   

    \begin{aligned} \displaystyle    
    \xi_{N}(X, Y) &\equiv 1 - \frac{\displaystyle 3\times \sum_{i=1}^{N-1}|r_{i+1}-r_i|}{N^2-1} \in [0, 1] \\ \\
    \end{aligned}  
  
    其中，$r_i$ 為 $y_i$ 在特徵變數 $Y$ 中的大小等級。    

  * 當兩特徵變數 $X$ 或 $Y$ 中存有重複的數據時，則，Chatterjee 相關係數 $\xi$ 的定義為  

    \begin{aligned} \displaystyle    
    \xi_{N}(X, Y) &\equiv 1 - \frac{\displaystyle N\times \sum_{i=1}^{N-1}|r_{i+1}-r_i|}{\displaystyle 2\times \sum_{i=1}^{N} l_i(N-l_i)} \in [0, 1] \\ \\
    \end{aligned}   

    其中，$r_i$ 為 $y_i$ 在特徵變數 $Y$ 中的大小等級，若存有多個 $y_i$ 相等重複，則，$r_i$ 為隨機均勻排序的結果，另外，$l_i$ 為在特徵變數 $Y$ 中大於 $y_i$ 的樣本總數。    

* <font color="#FFFF00">當 $\xi$ 越趨近於 1 時，表示兩特徵之間存有某種函數的對應關係，反之，**當 $\xi$ 越趨近於 0 時，表示兩特徵變數是獨立(Independent)**。</font>  

* 對於兩特徵變數的獨立性檢定，當數據平滑且非震盪時，$\xi$ 的功效似乎不如其他常見的檢定方法。
  > $\xi$ has only one disadvantage: It seems to have less power than several popular tests of independence when the signal is smooth and non-oscillatory.
    
    🔗 資料來源 : Chatterjee, S. (2021). A New Coefficient of Correlation. Journal of the American Statistical Association, 116(536), 2009–2022.   https://doi.org/10.1080/01621459.2020.1758115

````python
#@title Python 範例程式碼

#@markdown 下面範例為在三種不同的特徵關係下，Chatterjee 相關係數 $\xi$ 的大小。從下圖可知，Chatterjee 相關係數 $\xi$ 較上述傳統的相關係數(Pearson、Spearman 等)，能額外再捕捉出非線性(或單調)的關聯性。

import sys
import os
import random
import time

import numpy as np
import pandas as pd
import scipy
from scipy.stats import rankdata, norm
from collections import Counter
import matplotlib as mpl
import matplotlib.pyplot as plt

print('Python version:', sys.version[0:7])
print('NumPy version:', np.__version__)
print('Pandas version:', pd.__version__)
print('SciPy version:', scipy.__version__)
print('Matplotlib version:', mpl.__version__)
print('===================\n' )

# 定義函數 : 計算 Chatterjee 相關係數
def xicor(X, Y, ties=True, random_seed=42):
    """
    計算兩特徵變數的 Chatterjee 相關係數。

    params:
        X (array): 特徵變數的數據
        Y (array): 特徵變數的數據
        ties (boolean): 樣本數據中是否有重複的數值
        random_seed (int): 隨機種子數

    returns:
        statistic (float): 相關係數 Ksaai
        p_value (float): p value
    """

    # start_time = time.time()

    # 確認兩特徵欄位的資料個數是否相同
    if len(X)!=len(Y):
        raise IndexError(f"x, y length mismatch: {len(X)}, {len(Y)}")

    n = len(X)
    X_order = np.argsort(X)
    Y_after_sorting_X = Y[X_order]
    r = rankdata(Y_after_sorting_X, method='ordinal')
    nominator = np.sum(np.abs(np.diff(r)))

    # 當兩特徵欄位有重複的數值時
    if ties:
        random.seed(random_seed)

        # 分子
        for e in np.unique(Y_after_sorting_X):
            if len(np.where(Y_after_sorting_X==e)[0]) > 1:
                ties_Y_rank = r[np.where(Y_after_sorting_X==e)[0]]
                random.shuffle(ties_Y_rank)
                r[np.where(Y_after_sorting_X==e)[0]] = ties_Y_rank

        # print('r :', r)
        nominator = n*np.sum(np.abs(np.diff(r)))

        # 分母
        count_of_unique_value = list(Counter(Y).values())
        accu_count = np.cumsum(count_of_unique_value[::-1])[-2::-1]
        accu_count = list(np.append(accu_count, 0))
        denominator = 2*(sum((n-l)*l*c for l, c in zip(accu_count, count_of_unique_value)))

    # 當兩特徵欄位皆無重複的數值時
    else:
        nominator = 3*nominator
        denominator = np.power(n, 2)-1

    # 計算 Chatterjee 相關係數
    statistic = 1 - nominator/denominator
    p_value = norm.sf(statistic, scale=2 / 5 / np.sqrt(n))

    # print('--- Execution time = %s seconds ---' % (time.time()-start_time))
    return statistic, p_value

# 生成 3 組樣本數據
np.random.seed(42)
X = np.random.uniform(low=1, high=10, size=100)
Y1 = 3*X + 1.5*np.random.randn(100)
Y2 = X**2 + 30*np.random.randn(100)
Y3 = 0.8*np.sin(X) + 0.15*np.random.randn(100)

# 計算 Ksaai 相關係數
xi1, p1 = xicor(X, Y1)
xi2, p2 = xicor(X, Y2)
xi3, p3 = xicor(X, Y3)

# 資料視覺化
plt.figure(figsize=(12, 3))
plt.subplot(131)
plt.scatter(X, Y1, s=8)
plt.title('$\\xi$ = ' +str(np.round(xi1, 3)))
plt.ylabel('Y', rotation=0)

plt.subplot(132)
plt.scatter(X, Y2, s=8)
plt.title('$\\xi$ = ' +str(np.round(xi2, 3)))
plt.xlabel('X')

plt.subplot(133)
plt.scatter(X, Y3, s=8)
plt.title('$\\xi$ = ' +str(np.round(xi3, 3)))

img_path = 'Chatterjee_Correlation_Coefficient.png'
plt.savefig(img_path, bbox_inches='tight', format='png')

plt.show()
````

## 定義函數 : 計算 Chatterjee 相關係數

````python
def xicor(X, Y, ties=True, random_seed=42):
    """
    計算兩特徵變數的 Chatterjee 相關係數。

    params:
        X (array): 特徵變數的數據
        Y (array): 特徵變數的數據
        ties (boolean): 樣本數據中是否有重複的數值
        random_seed (int): 隨機種子數

    returns:
        statistic (float): Chatterjee 相關係數
        p_value (float): p value
    """

    # start_time = time.time()

    # 確認兩特徵欄位的資料個數是否相同
    if len(X)!=len(Y):
        raise IndexError(f"x, y length mismatch: {len(X)}, {len(Y)}")

    n = len(X)
    X_order = np.argsort(X)
    Y_after_sorting_X = Y[X_order]
    r = rankdata(Y_after_sorting_X, method='ordinal')
    nominator = np.sum(np.abs(np.diff(r)))

    # 當兩特徵欄位有重複的數值時
    if ties:
        random.seed(random_seed)

        # 分子
        for e in np.unique(Y_after_sorting_X):
            if len(np.where(Y_after_sorting_X==e)[0]) > 1:
                ties_Y_rank = r[np.where(Y_after_sorting_X==e)[0]]
                random.shuffle(ties_Y_rank)
                r[np.where(Y_after_sorting_X==e)[0]] = ties_Y_rank

        # print('r :', r)
        nominator = n*np.sum(np.abs(np.diff(r)))

        # 分母
        count_of_unique_value = list(Counter(Y).values())
        accu_count = np.cumsum(count_of_unique_value[::-1])[-2::-1]
        accu_count = list(np.append(accu_count, 0))
        denominator = 2*(sum((n-l)*l*c for l, c in zip(accu_count, count_of_unique_value)))

    # 當兩特徵欄位皆無重複的數值時
    else:
        nominator = 3*nominator
        denominator = np.power(n, 2)-1

    # 計算 Chatterjee 相關係數
    statistic = 1 - nominator/denominator
    p_value = norm.sf(statistic, scale=2 / 5 / np.sqrt(n))

    # print('--- Execution time = %s seconds ---' % (time.time()-start_time))
    return statistic, p_value
````
